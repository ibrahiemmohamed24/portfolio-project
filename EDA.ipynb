```python
#bringing in our data
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r"D:\New folder\student_scores (1).csv")
df = df.drop("Unnamed: 0",axis=1)
df.head()
```

```python
#Lets bring the English scores
df2 = pd.read_csv(r"D:\New folder\English_grades.csv")
df2 = df2.drop("Unnamed: 0",axis=1)
df2.head()
```

```python
df = df.merge(df2, on="Name")
df.head()
```

```python
#checking the data frame info

df.info()
```

```python
#filling null values

avg_age = df["Age"].mean()
df["Age"] = df["Age"].fillna(avg_age)
df["Math Score"] = df["Math Score"].fillna(0)
df.info()
```

```python
#unique values
df["City"].unique()
```

```python
#replacing values and stings
df["City"] = df["City"].str.replace("Port", "pt.")
df["City"].unique()
```

```python
#replacing values and value count
df["Gender"] = df["Gender"].replace({'Male':'M', 'Female':'F', 'Other':'0'})
df["Gender"].value_counts(normalize=True)
```

```python
#grouping function
#df.groupby('Gender')['Math Score'].sum()
df.groupby('Gender')['Math Score'].agg(['sum', 'mean'])
```

```python
#pivot table
import numpy as np
pd.pivot_table(df, values = ['Math Score', 'Science Score', 'English Score'], index='Gender', aggfunc=np.mean, margins=True)
```

```python
#Transform
df['Total Score'] = df['Math Score'] + df['English Score'] + df['Science Score']
df['Average per Group'] = df.groupby('Gender')['Total Score'].transform('mean')
df.head()
```

```python
#numpy conditional
```

```python
avg_math = df['Math Score'].mean()
df['Above Average Math Score'] = np.where(df['Math Score'] >avg_math, 'Yes', 'No')
df['Above Average Math Score'].value_counts(normalize=True)
```
